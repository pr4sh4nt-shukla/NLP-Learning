```python
corpus = """ Hello Welcom, I'm Prashant who's learning NLP.
I'm Very Exitecd! I Hope i'll learn and give my best.
"""
```

```python
print(corpus)
```

## Tokenization
## Sentence --> paragraphs


```python
from nltk.tokenize import sent_tokenize
```

```python
document = sent_tokenize(corpus)
```

```python
 document
```

```python
type(document)
```

```python
for sentence in document:
    print(sentence)
```

## Tokenization
## Paragraph --> words
## Sentence --> Words

```python
from nltk.tokenize import word_tokenize
# it conver whole sentece to word without punction 
```

```python
word= word_tokenize(corpus)
```

```python
word
```

```python
for sentence in word:
    print(sentence)
```

```python
from nltk.tokenize import wordpunct_tokenize
# it can convert sentence to single word even punction we can see that " ' " is seperated from ''
```

```python
punt_tokenize = wordpunct_tokenize(corpus)
```

```python
punt_tokenize
```

```python
from nltk.tokenize import TreebankWordTokenizer
```

```python
tokenizer = TreebankWordTokenizer()
```

```python
tokenizer.tokenize(corpus)
```
