# Wordnet Lemmetizer
Lemmetization is a technique like  stemming. The Output we'll get after lemmitiztion is called 'Lemma' which is a root word rather than root stem, the output of stemming. After lemmetization we'll getting the valid word that has means the same thing.

Nltk Provides WordNetLemmetizer class which is a thin wrapper around wordnet corpus. This class used morphy() function to WordNet CorpusReader class to find a lemma. Let us understand with a example:-


```python
## Q&A , chatbox , text summarization which is best exampel of lemmetization
from nltk.stem import WordNetLemmatizer
```

```python
import nltk
nltk.download('wordnet')
```

```python
lemmatizer = WordNetLemmatizer()
```

it used a "Pos" keyword which explain  which kind or word that is.
* Noun - n
* Verb - v
* Adjective - a
* Adverb - r

```python
# this is same words we used in stemming to see what is differecne from stemming
words = ["eating","eat","eaten","writing","writes","programming","programs","history","finally","finalize"]
```

```python
for word in words:
    print(f"{word} ---> {lemmatizer.lemmatize(word,pos='v')} ")
```

```python
lemmatizer.lemmatize("Congratulation",pos='v')
```

```python
lemmatizer.lemmatize('fairly',pos='r'), lemmatizer.lemmatize('automatically',pos='v')
```
