# Stemming
*Stemming is the process of reducing a word to its word stem that affixes to suffixes and prefix or to the roots of word known as a lemma. Stemming is important is natural language understanding (NLU) and natural language processing (NLP).**

```python
## classification Problem
## comments of products is a positive review or negetive review.
## Review--> eating , eat, eaten or going , go , goes
words = ["eating","eat","eaten","writing","writes","programming","programs","history","finally","finalize"]
```

# PorterStemmer

```python
from nltk.stem import PorterStemmer
```

```python
stemming = PorterStemmer()
```

```python
for word in words:
    print(f" {word} -----> {stemming.stem(word)}")
```

```python
stemming.stem("Congratulation")
```

# RegexStemmer

Nltk has RegexStemmer class with the help of which we can easily implement Regular Expression Stemmer Algorithms. It Basically takes a single regular expression and removes any prefix or suffix that matches the expression

```python
from nltk.stem import RegexpStemmer
```

```python
regex_stemmer = RegexpStemmer("ing$|s$|able$|tion$", min=4)
```

```python
for word in words:
    print(f"{word}----> {regex_stemmer.stem(word)}")
```

if we write "ing" first in starting of word it removes last one cuz we put "$" in the end of suffix. For Example:-

```python
regex_stemmer.stem('inggoing')
```

```python
regex_stemmer = RegexpStemmer("ing|s$|able$|tion$", min=4) # i removed last dollar now it removes "ing" no matter where it is.
```

```python
regex_stemmer.stem('inggoing'), regex_stemmer.stem('going'), regex_stemmer.stem('eating_food')
```

# SnowBallStemmer

The Snowball Stemmer, also known as the Porter2 stemmer, is a widely used and efficient stemming algorithm in Natural Language Processing (NLP) that reduces words to their base or root forms (stems). It was developed by Martin Porter as an improvement over his original Porter stemmer

```python
from nltk.stem import SnowballStemmer
```

```python
snow_stemmer = SnowballStemmer('english')
```

```python
for word in words:
    print(f"{word}-----> {snow_stemmer.stem(word)}")
```

we can see that it's almost similar as "PorterStemmer" for same words that we applied to both stemmer. but SnowBallStemmer work better then PorterStemmer. for example:-

```python
stemming.stem('fairly'), stemming.stem('automatically')
```

```python
snow_stemmer.stem('fairly'), snow_stemmer.stem('sportly')
```
